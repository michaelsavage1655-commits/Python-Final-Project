```python
#data visualization

def c(a,b):
  pa = pro[sct.index(a)]
  pb = pro[sct.index(b)]
  sns.set_style("whitegrid")
  sns.set_context("notebook", font_scale=1.2)
  fig, ax = plt.subplots(figsize=(10, 6))
  sns.scatterplot(
      data=df,
      x=a,
      y=b,
      ax=ax,
      color='#1f77b4',
      alpha=0.7,
      edgecolor='k',
      linewidth=0,
      s=60
  )

  ax.set_title(pa.capitalize()+" vs. "+pb.capitalize(), fontsize=16, pad=15)
  ax.set_xlabel(pa.capitalize(), fontsize=13)
  ax.set_ylabel(pb.capitalize(), fontsize=13)


  plt.tight_layout()
  plt.show()



# CHARTS GALORE!!!
sct = ['time', 'released_year', 'age', 'nb_of_runs', 'TAS_time', 'time_difference', '%_of_difference']

pro = ['Time','Released Year','Age','Number Of Runs','Tas Time','Time Difference','Percent Of Difference']
# This piece of code auto generates every possible scatterplot between numeric variables.
for q,w in (list(combinations(sct, 2))):
  c(q,w)

```
